```python
import random
import time

short_list = [6,5,3,1,8,7,2,4]
long_list = random.sample(range(1000000), 10000)
```

```python
# Our merge function takes two ordered lists and merges them together into one ordered list

def merge(a, b):
    # Check for empty list.
    if len(a) == 0 or len(b) == 0:
        return a or b
    
    # Start with an empty result.
    result = []
    # Track two indexes.
    i, j = 0, 0
    # Set a while condition to ensure we iterate only for the length of our two lists.
    while (len(result) < len(a) + len(b)):
        # If a's next element is lower append that element to our result.
        if a[i] < b[j]:
            result.append(a[i])
            i += 1
        # Otherwise append b's next element.
        else:
            result.append(b[j])
            j += 1
        # When one list is empty just append everything from the other list and stop.
        if i == len(a) or j == len(b):
            result.extend(a[i:] or b[j:])
            break 

    return result

def merge_sort(lst):
    if len(lst) < 2:
        return lst

    mid = int(len(lst) / 2)
    a = merge_sort(lst[:mid])
    b = merge_sort(lst[mid:])

    return merge(a, b)
```

```python
## Run and measure runtime for short list
start_time = time.time()
new_list = merge_sort(short_list)
print('Runtime: {} seconds'.format(time.time()-start_time))
print('Original list: {}'.format(short_list))
print('Ordered list:  {}'.format(new_list))
```

```python
## Run and measure runtime for long list
start_time = time.time()
new_list = merge_sort(long_list)
print('Runtime: {} seconds'.format(time.time()-start_time))
print('Original list: {}'.format(long_list))
print('Ordered list:  {}'.format(new_list))
```
