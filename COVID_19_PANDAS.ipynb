```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from jupyterthemes import jtplot
import warnings

jtplot.style(theme="monokai", context="notebook", ticks=True, grid=False)
sns.set_style("whitegrid")

df_confd = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
df_death = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
```

```python
# Reshape confirmed
df_confd = df_confd[df_confd['Province/State'].isnull() == True]
df_confd = df_confd.drop(columns = ['Lat','Long', 'Province/State'])
df_confd = df_confd.T
df_confd.columns = df_confd.iloc[0]
df_confd = df_confd.drop('Country/Region', axis=0)
df_confd = df_confd.reset_index().rename(columns = {'index':'Date'})
df_confd['Date'] = pd.to_datetime(df_confd['Date'], utc=False)
df_confd = df_confd.set_index('Date')
df_confd.index = df_confd.index.date
df_confd = df_confd.iloc[34:]
```

```python
# Reshape deaths
df_death = df_death[df_death['Province/State'].isnull() == True]
df_death = df_death.drop(columns = ['Lat','Long', 'Province/State'])
df_death = df_death.T
df_death.columns = df_death.iloc[0]
df_death = df_death.drop('Country/Region', axis=0)
df_death = df_death.reset_index().rename(columns = {'index':'Date'})
df_death['Date'] = pd.to_datetime(df_death['Date'], utc=False)
df_death = df_death.set_index('Date')
df_death.index = df_death.index.date
df_death = df_death.iloc[34:]
```

```python
fig = plt.figure(figsize=(20,10))
fig, axes = plt.subplots(nrows=1, ncols=2)
#country_list = ['US','Italy', 'Spain', 'Germany', 'Korea, South', 'Japan']
country_list = ['Korea, South', 'Japan']
df_confd[country_list].plot(kind = 'bar', figsize=(30,8), stacked=True, title = 'COVID-19: Confirmed Cases', ax=axes[0])
df_death[country_list].plot(kind = 'bar', figsize=(30,8), stacked=True, title='COVID-19: Associated Deaths', ax=axes[1])
```
