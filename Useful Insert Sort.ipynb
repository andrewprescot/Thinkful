```python
import random
import time

short_list = [6,5,3,1,8,7,2,4]
long_list = random.sample(range(1000000), 10000)
```

```python
## Insert Sort

def insert_sort(data):
    
    new_list = data.copy()
    
    for ii in range(len(new_list)):
        
        jj = ii
        
        while jj>0 and new_list[jj-1]>new_list[jj]:
            # Swaps elements
            new_list[jj-1], new_list[jj] = new_list[jj], new_list[jj-1]
            jj-=1
            
    return new_list
```

```python
## Run and measure runtime for short list
start_time = time.time()
new_list = insert_sort(short_list)
print('Runtime: {} seconds'.format(time.time()-start_time))
print('Original list: {}'.format(short_list))
print('Ordered list:  {}'.format(new_list))
```

```python
## Run and measure runtime for long list
start_time = time.time()
new_list = insert_sort(long_list)
print('Runtime: {} seconds'.format(time.time()-start_time))
print('Original list: {}'.format(long_list))
print('Ordered list:  {}'.format(new_list))
```
