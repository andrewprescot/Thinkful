# INTRODUCTION<br/>
Recent epidemiological studies have identified (i) altitude and (ii) specific airborne pollutants as independent risk factors for suicide and/or suicidal behavior. Specifically, nitrogen dioxide **(NO<sub>2</sub>)**, which is formed primarily through the burning of fossil fuels, was found to be associated with an increased odds ratio of 20% in suicide completers.<br/>

**Salt Lake City (SLC)**, Utah, is an urban agglomeration situated at approximately 4,500 ft above sea level. It is also noteworthy that SLC recently was identified as the nation's 7th worst air quality. SLC is surrounded by the Oquirrh mountain to the west and the Wastch mountain range to the east, with the two mountain ranges meeting at the south. During the colder fall and winter months an interplay of natural meterological events act to form a layer of warm air, which effectively traps colder air and its pollutants within the SLC valley atmosphere. Hence, due to this **inversion effect**, SLC's air quality is significantly deterioated during fall and winter.<br/>

In contrast, **Honolulu (HON)**, Hawaii, is a similarly sized urban area of residence (HON vs. SLC population: ~300,000 vs 200,000) with an overall air quality that is ranked amongst the nation's best. HON is situated at sea level and thus **does not** experience seasonal inversion effects.<br/>



The United States (US) Clean Air Act (est. ~1970) requires the US Environment Protection Agency (EPA) to set air quality standards for six common air pollutants, including NO<sub>2</sub>. The National Ambient Air Quality Standard (NAAQS) for NO<sub>2</sub> measures currently is set to a daily one-hour maximum of 100 parts per billion (ppb), and the national average for air NO<sub>2</sub> concentration has decreased substantially over the past couple of decades. The EPA releases outdoor air quality data for several airborn pollutants across the US, including the **daily one-hour maximum** NO<sub>2</sub> measures for SLC and HON.<br/>


Given the implication of NO<sub>2</sub> in suicidal behavior in at-risk clinical populations, a team of psychiatrists at the University of Utah are keen to investigate the daily one-hour maximum NO<sub>2</sub> levels for SLC and HON over a two-decade timespan, including the years 1999, 2004, 2009, 2014, 2018, and 2019. They are also interested in seasonal effects and the potential increase of NO<sub>2</sub> levels during the inversion months (September through February) within SLC from 2018 to 2019.

**_Specific research questions include:_**

1. Is there a significant decrease in SLC atmospheric NO<sub>2</sub> concentration from 1999 to 2019?
2. Is there a significant decrease in HON atmospheric NO<sub>2</sub> concentration from 1999 to 2019?
3. Are SLC atmospheric NO<sub>2</sub> concentrations significantly different from correposnding HON measures between 1999 and 2019?
4. Is SLC atmospheric NO<sub>2</sub> concentration higher during fall and winter months as compared to spring and summer months? 
5. Is HON atmospheric NO<sub>2</sub> concentration higher during fall and winter months as compared to spring and summer months? 
<br/>

Data source: <https://www.epa.gov/outdoor-air-quality-data/download-daily-data>

# DATA<BR/>

All data were obtained from the publicly-accessible US EPA respository for outdoor data air quality (link to download: <https://www.epa.gov/outdoor-air-quality-data/download-daily-data>), which provides a tool that queries daily air quality summaries for a given **pollutant**, **year**, **geographic US location**, and **monitor site**. For most locations, data can be downloaded from multiple monitor sites. The two monitor sites deemed most representative for urban SLC and urban HON are the **Hawthorne** and **Kapolei** stations, respectively.<br/>

Daily one-hour maximum SLC and HON NO<sub>2</sub> data in the form of comma separated value (CSV) files subsequently were individually downloaded for years 1999, 2004, 2009, 2014, 2018, and 2019. The individual CSV files were concatenated into a single dataframe using the Pandas libray (Python 3.7.4), which enabled the assembly of a master composite CSV file. Two addtional variables were appended to the file that enhanced usability, including 'year' (four digit integer) and 'site' (i.e. 'SLC' for 'Hawthorne', and 'HON' for 'Kapolei'). This resulted in a master CSV file containing 4056 observations and 22 variables. The SLC and HON sites showed a total of 64 days and 260 days of missing data, respectively. The code used to build the master CSV file is provided in the cell below.

```python
## Create Master CSV File: 'create_master_no2' flag set to False by default. 
create_master_no2 = False
import pandas as pd

if create_master_no2:  
    base = 'https://raw.githubusercontent.com/aprescot1977/Thinkful/master/'
    
    # Salt Lake City 1999, 2004, 2009, 2014, 2018, 2019 csv data read-in
    # and concatenation to master SLC

    slc_no2_1999 = pd.read_csv(base+'slc_1999_no2.csv', parse_dates = ['Date'])
    slc_no2_1999['year']= 1999
    slc_no2_2004 = pd.read_csv(base+'slc_2004_no2.csv', parse_dates = ['Date'])
    slc_no2_2004['year']= 2004
    slc_no2_2009 = pd.read_csv(base+'slc_2009_no2.csv', parse_dates = ['Date'])
    slc_no2_2009['year']= 2009
    slc_no2_2014 = pd.read_csv(base+'slc_2014_no2.csv', parse_dates = ['Date'])
    slc_no2_2014['year']= 2014
    slc_no2_2018 = pd.read_csv(base+'slc_2018_no2.csv', parse_dates = ['Date'])
    slc_no2_2018['year']= 2018
    slc_no2_2019 = pd.read_csv(base+'slc_2019_no2.csv', parse_dates = ['Date'])
    slc_no2_2019['year']= 2019
    slc_no2_master = pd.concat([slc_no2_1999, slc_no2_2004, slc_no2_2009, slc_no2_2014, slc_no2_2018, slc_no2_2019])

    slc_no2_master = slc_no2_master[slc_no2_master['Site Name'] == 'Hawthorne']
    slc_no2_master['site'] = 'SLC'

    # Honolulu Lake City 1999, 2004, 2009, 2014, 2018, 2019 csv data read-in
    # and concatenation to master HON

    hon_no2_1999 = pd.read_csv(base+'hi_1999_no2.csv', parse_dates = ['Date'])
    hon_no2_1999['year']= 1999
    hon_no2_2004 = pd.read_csv(base+'hi_2004_no2.csv', parse_dates = ['Date'])
    hon_no2_2004['year']= 2004
    hon_no2_2009 = pd.read_csv(base+'hi_2009_no2.csv', parse_dates = ['Date'])
    hon_no2_2009['year']= 2009
    hon_no2_2014 = pd.read_csv(base+'hi_2014_no2.csv', parse_dates = ['Date'])
    hon_no2_2014['year']= 2014
    hon_no2_2018 = pd.read_csv(base+'hi_2018_no2.csv', parse_dates = ['Date'])
    hon_no2_2018['year']= 2018
    hon_no2_2019 = pd.read_csv(base+'hi_2019_no2.csv', parse_dates = ['Date'])
    hon_no2_2019['year']= 2019
    hon_no2_master = pd.concat([hon_no2_1999, hon_no2_2004, hon_no2_2009, hon_no2_2014, hon_no2_2018, hon_no2_2019])

    hon_no2_master = hon_no2_master[hon_no2_master['Site Name'] == 'Kapolei']
    hon_no2_master['site'] = 'HON'

    # Generate a two-site master csv file for SLC and HON
    Prescot_Thinkful_C1_NO2_master = pd.concat([slc_no2_master, hon_no2_master])
    Prescot_Thinkful_C1_NO2_master.to_csv('D:\\D_DRIVE_BACKUP2016\\DataScience\\capstone1\\Prescot_Thinkful_C1_NO2_master.csv',  index=False, header=True) # save locally
    Prescot_Thinkful_C1_NO2_master.info()
```

# METHODS<BR/>

**Hypothesis 1 and 2** will be tested using the main dataframe (i.e. from 'Prescot_Thinkful_C1_NO2_master.csv'.csv'). The normaliy of the atmospheric NO<sub>2</sub>  concentration distributions for years 1999, 2004, 2009, 2014, 2018, and 2019, will be evaluated using histogram analysis, as well as measures of skewness and kurtosis. In addition, Shapiro-Wilk tests will likely be perfomed. The 'site' variable will be used to differentiate the 'SLC' from 'HON' distributions and for the resulting within-site statistical analysis. If the SLC and HON atmospheric NO<sub>2</sub>  concentration distributions are normally distributed, a one-way ANOVA will be performed, otherwise the data will be treated using non-parametric tests.
___
**Hypothesis 3** also will be tested using the main dataframe (i.e. from 'Prescot_Thinkful_C1_NO2_master.csv'.csv'). Normality for the 6 discrete annual distributions will already have been conducted as detailed above. However, the annual SLC versus HON atmospheric NO<sub>2</sub>  concentrations will be statistically compared on a year-by-year basis. The use of parametric verus non-parametric statistical analysis will be dependent on normality of SLC and HON NO<sub>2</sub> distributions.
___
**Hypothesis 4 and 5** will be tested by creating a separate dataframe. The 'Date' column of the main dataframe (i.e. from 'Prescot_Thinkful_C1_NO2_master.csv'.csv') will be used to create a separate dataframe depending on whether the days fall within fall 2018, winter 2018, spring 2019, summer 2019, and fall 2019. These are defined as below:<br/><br/>


* 2018-09-01 to 2018-11-30: **Fall 2018:**   
* 2018-12-01 to 2019-02-28: **Winter 2018:** 
* 2019-03-01 to 2019-05-31: **Spring 2019:** 
* 2019-06-01 to 2019-08-31: **Summer 2019:** 
* 2019-09-01 to 2019-11-30: **Fall 2019:**<br/><br/>

These specific dates will be accessed using the Pandas **between()** method, and a new column will be generated that is allocated a string showing the relevant 'season' and 'year'. The normality of the 5 new seasonal distrubutions (each containing approximately 90 observations) will be tested for both SLC and HON using the methods outlined above (see Hypothesis 1 and 2). The seasonal variation of SLC and HON air NO<sub>2</sub> concentration will be tested using parametric or non-parametric statistical analyis, depending on the outcome of the distribution analysis.

### Summary of distributions<br/>

Therefore, non-parametric statistical methods will be used to test all within-site and between-site hypotheses (Hypotheses 1-3).

## Test Hypothesis 1<br/>

**Ho:** there is no significant decrease in SLC atmospheric NO<sub>2</sub> concentrations from 1999 to 2019<br/><br/>

Data analysis steps:<br/>
___
(i) Kruskal-Wallis H-test (one-way ANOVA on ranks)<br/> 
(ii) Post hoc test for multiple comparisons of mean rank sums (Dunn's test using the Bonferroni method for adjustment of p values)<br/>
(iii) Explicit median calculation for each of the 6 year periods<br/>
(iv) Barplot visualization using Seaborn<br/>

```python
# (i) Kruskal-Wallis for SLC
print(stats.kruskal(df[(df['year']==1999)&(df['site']=='SLC')][pollutant_no2],
                    df[(df['year']==2004)&(df['site']=='SLC')][pollutant_no2], 
                    df[(df['year']==2009)&(df['site']=='SLC')][pollutant_no2], 
                    df[(df['year']==2014)&(df['site']=='SLC')][pollutant_no2],
                    df[(df['year']==2018)&(df['site']=='SLC')][pollutant_no2],
                    df[(df['year']==2019)&(df['site']=='SLC')][pollutant_no2]))
```

```python
# (ii) Dunn's test for SLC
import scikit_posthocs as sp
sp.posthoc_dunn(df[df['site']=='SLC'], val_col = pollutant_no2, group_col = 'year', p_adjust = 'bonferroni')
```

```python
# (iii) Median calculation for SLC
[print('SLC {} NO2 median: {}'.format(ii, df[(df['year']==ii)&(df['site']=='SLC')][pollutant_no2].median())) for ii in years];
```

```python
# (iv) Barplot visualization for SLC (1999-2019)
fig, axes = plt.subplots(1,1,figsize = (20, 10))
ax = sns.boxplot(y = pollutant_no2, x = 'year', data = df[df['site']=='SLC'])
ax.axhline(100, color = 'k', linestyle='dashed');
ax.set_title('Salt Lake City ' + pollutant_no2 + ' for 6 separate years');
style = dict(size=15, color='k');
ax.text(3, 102, "NAAQS One Hour Maximum Standard Level", **style);
```

### Hypothesis 1: summary<br/>

* Based on Kruskal-Wallis test, we reject the null hypothesis that there is no decrease in median daily one-hour maximum NO<sub>2</sub> concentration in SLC from 1999 to 2019 (p<0.05). The median daily one-hour maximum NO<sub>2</sub> concentration in 1999 was 43 ppb compared to 26 ppm in 2019.<br/><br/>

* Subsequent post-hoc testing (Dunn-Bonferroni) allowed for a better undertsanding of statistical relationships. For example, the median NO<sub>2</sub> concentration for 1999 was significantly higher than the corresponding value fo all other years (p<0.05). However, the SLC NO<sub>2</sub> concentration for 2019 did not differ significantly from the corresponding 2014 and 2018 measures.

## Test Hypothesis 3<br/>

**Ho:** there is no significant difference between SLC and HON atmospheric NO<sub>2</sub> concentrations between 1999 and 2019<br/><br/>
___
(i) Mann-Whitney U-test for each of the 6 year periods<br/>
(ii) Barplot visualization using Seaborn<br/>

```python
# (ii) plot SLC and HON levels for all years, x = 'year' and hue = 'site'
# Slightly different visualization that more closely represents the 6 individual Mann-Whitney U-tests
fig, axes = plt.subplots(1,1,figsize = (20, 10))
ax = sns.boxplot(y = pollutant_no2, x = 'year', data = df, hue = 'site')
ax.axhline(100, color = 'k', linestyle='dashed');
ax.set_title('Salt Lake City and Honolulu ' + pollutant_no2 + ' for 6 separate years');
style = dict(size=15, color='k');
ax.text(0.75, 102, "NAAQS One Hour Maximum Standard Level", **style);
```

### Hypothesis 3: summary<br/>

* As the SLC NO2 concentration steadily decreases from 1999 to 2019, so to do does the magnitude of statistical significance of the SLC-to-HON NO<sub>2</sub> comparisons.

___

(i) Create a seasonal fall 2018 to fall 2019 dataframe<br/>
(ii) Analysis of distributions for both SLC and HON (histograms, QQ plots, statistics)

```python
# (i) Seasonal effects: first build a subset of data spanning autumn 2018 to autumn 2019 from master data frame
df_spring18 = pd.DataFrame(df.loc[df['Date'].between('2018-03-01','2018-05-31')])
df_summer18 = pd.DataFrame(df.loc[df['Date'].between('2018-06-01','2018-08-31')])
df_autumn18 = pd.DataFrame(df.loc[df['Date'].between('2018-09-01','2018-11-30')])
df_winter18 = pd.DataFrame(df.loc[df['Date'].between('2018-12-01','2019-02-28')])
df_spring19 = pd.DataFrame(df.loc[df['Date'].between('2019-03-01','2019-05-31')])
df_summer19 = pd.DataFrame(df.loc[df['Date'].between('2019-06-01','2019-08-31')])
df_autumn19 = pd.DataFrame(df.loc[df['Date'].between('2019-09-01','2019-11-30')])

df_spring18['season'] = 'spring 2018'
df_summer18['season'] = 'summer 2018'
df_autumn18['season'] = 'autumn 2018'
df_winter18['season'] = 'winter 2018'
df_spring19['season'] = 'spring 2019'
df_summer19['season'] = 'summer 2019'
df_autumn19['season'] = 'autumn 2019'

df_season = pd.concat([df_spring18, df_summer18, df_autumn18, df_winter18, df_spring19, df_summer19, df_autumn19])
df_season.info()
```

## Test Hypothesis 4<br/>

**Ho:** there is no significant difference in SLC atmospheric NO<sub>2</sub> concentration in fall and winter months compared to spring and summer months<br/><br/>

Data analysis steps:<br/>
___
(i) Kruskal-Wallis H-test (one-way ANOVA on ranks)<br/>
(ii) Post hoc test for multiple comparisons of mean rank sums (Dunn's test using the Bonferroni method for adjustment of p values)<br/>
(iii) Explicit median calculation for each of the five separate 3-month periods<br/>
(iv) Barplot visualization using Seaborn<br/>

```python
# (i) Seasonal statistics to compare SLC NO2 level throughout the year
print(stats.kruskal(df_season[(df_season['season']=='autumn 2018')&(df_season['site']=='SLC')][pollutant_no2],
                    df_season[(df_season['season']=='winter 2018')&(df_season['site']=='SLC')][pollutant_no2],
                    df_season[(df_season['season']=='spring 2019')&(df_season['site']=='SLC')][pollutant_no2],
                    df_season[(df_season['season']=='summer 2019')&(df_season['site']=='SLC')][pollutant_no2],
                    df_season[(df_season['season']=='autumn 2019')&(df_season['site']=='SLC')][pollutant_no2]))
```

```python
# (ii) Dunn's test for SLC NO2 level throughout the year
sp.posthoc_dunn(df_season[df_season['site']=='SLC'], val_col = pollutant_no2, group_col = 'season', p_adjust = 'bonferroni')
```

```python
# (iii) Medians for SLC NO2 level throughout the year
[print('{} SLC seasonal NO2 median: {}'.format(ii, df_season[(df_season['season']==ii)&(df_season['site']=='SLC')][pollutant_no2].median())) for ii in season_list];
```

```python
# (iv) Barplot visualization of SLC NO2 throughout the year
fig, axes = plt.subplots(1,1,figsize = (20, 10))
ax = sns.boxplot(y = pollutant_no2, x = 'season', data = df_season[df_season['site']=='SLC'])
ax.axhline(100, color = 'k', linestyle='dashed');
ax.set_title('Salt Lake City ' + pollutant_no2 + ': Autumn 2018 through Autumn 2019');
style = dict(size=15, color='k');
ax.text(0.75, 102, "NAAQS One Hour Maximum Standard Level", **style);
```

### Hypothesis 4: summary<br/>

* Based on the Kruskal-Wallis H-tests, we reject the null hypothesis that there is no significant difference in SLC atmospheric NO<sub>2</sub> concentration in fall and winter months compared to spring and summer months (p<0.05).<br/><br/>

* Subsequent post-hoc testing (Dunn-Bonferroni) allowed for a better undertsanding of statistical relationships. For example, the median SLC NO<sub>2</sub> concentration for Winter 2018 (40.0 ppb) was significantly higher than the corresponding value for summer 2019 (17.0 ppm). Notably, from summer 2019 to autumn 2019, there was a significant change in NO<sub>2</sub> concentration with MEDIAN levels increasing from 17.0 to 33.4 ppm (p<0.05).

# DISCUSSION AND RECOMMENDATIONS<BR/>

* In general, SLC atmospheric NO<sub>2</sub> concentrations were found to be below the NAAQS one hour maximum standard level of 100 ppb. From 1999 to 2019, there was a statistically significant (almost twofold) reduction in measured SLC NO<sub>2</sub> levels with the median falling from 48 ppb to 26 ppm.<br/><br/>

* HON atmospheric NO<sub>2</sub> concentrations were found to be well below the NAAQS one hour maximum standard level of 100 ppb. From 1999 to 2019, there was a smaller yet signficant (~25%) reduction in measured HON NO<sub>2</sub> levels with the median falling from 12.0 ppb to 8.8 ppm.<br/><br/>

* For the six separate years under investigation (1999-2019), SLC showed statistically significant higher NO<sub>2</sub> concentrations than the corresponding HON measures. In 1999, the median SLC NO<sub>2</sub> measure was approximately **4 times greater** than the equivalent HON measure. In 2019, the median SLC NO<sub>2</sub> measure was approximately **3 times greater** than the equivalent HON measure. <br/><br/>

* For SLC, significant changes in NO<sub>2</sub> levels were detected depending on the season under investigation. For example, the median NO<sub>2</sub> concentration measured for winter 2018 (40.0 ppb) was over double the corresponding summer 2019 measure (17.0 ppb). Although HON NO<sub>2</sub> measures showed a significant decrease from winter 2018 to summer 2019, the magnitude of the effect was much smaller (~30% reduction).
